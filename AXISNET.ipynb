```python
from google_play_scraper import Sort, reviews_all, reviews
import pandas as pd
```

```python
result, continuation_token = reviews(
    'com.axis.net',          # Package ID AXISnet
    lang='id',               # Bahasa Indonesia
    country='id',            # Negara Indonesia
    sort=Sort.NEWEST,        # Ambil ulasan terbaru
    count=1500               # Jumlah ulasan yang mau diambil
)
```

```python
df_bus = pd.DataFrame(result)
df_bus.to_csv("hasil_scraping_axisnet.csv", index=False)
print("Scraping selesai! File hasil_scraping_axisnet.csv berhasil disimpan.")
```
